# **Pre-Processing**

```python
 df=pd.get_dummies(titanic)
 df
```

# **Apriori algorithm**

```python
frequent_items=apriori(df,min_support=0.1,use_colnames=True) # min_support default value is 0.5
#min_support is a float b/w 0 and 1 for minimum support of the itemsetsreturned
# the support is computed as the fraction transactions_where_item(s)_occur / total_transactions
frequent_items # o/p is frequent itemsets: 1-item itemsets,2-item itemsets etc.with min-support criteria
```

```python
rules=association_rules(frequent_items,metric="lift",min_threshold=0.7)# min_threshold is confidence
rules
rules.sort_values("lift",ascending=False)[0:20]
```

```python
rules[rules.lift>1]
```
