## **MODEL Building**

```python
x=df.iloc[:,:-1]
y=df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
ypred
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr.score(x_train,y_train),lr.score(x_test,y_test)
```

## **1.Lasso (L1) Regression**

```python
#after lasso, train accuracy will decrease but test accuracy will increase
l1=Lasso()
l1.fit(x_train,y_train)
l1.score(x_train,y_train),l1.score(x_test,y_test)
 # we got the same accuracy and ooerfitted model
```

```python
l1.coef_
```

### **hyperparameter tuning**

```python
for i in range(100,200):
  l1=Lasso()
  l1.fit(x_train,y_train)
  print(f"Alpha: {i} Train:{l1.score(x_train,y_train)} Test:{l1.score(x_test,y_test)}")
```

```python
# as alpha increases train accuracy decreases and test accuracy increases
# so try big alpha value
l1=Lasso(alpha=175)
l1.fit(x_train,y_train)
l1.score(x_train,y_train),l1.score(x_test,y_test)
```

```python
l1.coef_
```

```python
x_train.columns
```

## **Ridge (L2) regression**

```python
l2=Ridge()
l2.fit(x_train,y_train)
l2.score(x_train,y_train),l2.score(x_test,y_test)
```

```python
for i in range(1,100):
  l2=Ridge()
  l2.fit(x_train,y_train)
  print(f"Alpha: {i} Train:{l2.score(x_train,y_train)} Test:{l2.score(x_test,y_test)}")
```

```python
l2=Ridge(alpha=8)
l2.fit(x_train,y_train)
l2.score(x_train,y_train),l2.score(x_test,y_test)
```

```python
l2.coef_
```

## **3.Elastic Net**

```python
l3=ElasticNet(alpha=8)
l3.fit(x_train,y_train)
l3.score(x_train,y_train),l3.score(x_test,y_test)
```
