## **Isolation Forest**

```python
# isolation forest example - useful for anamoly detection along with outliers
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
df=pd.DataFrame({
    'marks':[50,70,67,80,90],
    'Time':[1,2,3,1.5,60]
})
df
```

```python
# what is outlier here?
# iso=OsolationForest(random_state=10) expecting 10% outliers in dataset
iso=IsolationForest(random_state=10,contamination=0.1)
iso.fit(df)
out=iso.predict(df)
```

```python
out
```

```python
df['score']=out
df
```

```python
data_enc
```

```python
# training the model
cif=IsolationForest(random_state=10,contamination=.01)
# contamination: how much percentage of outliers you are expecting in the dataset
# eg. in health care domain it will be very low(1% or less)0.01 or0.001
cif.fit(data_enc)
```

```python
# predictions
y_pred_outliers=cif.predict(data_enc)
```

```python
y_pred_outliers
```

```python
import ppscore as pps
import pandas as pd
```

```python
df1=pd.read_csv("/content/Iris.csv")
df1
```

```python
# syntax: pps.score(df,'feature_col','target_col')
pps.score(df1,'SepalLengthCm','PetalLengthCm')
```

```python
# calculate the whole pps matrix
pps.matrix(df1)
```

```python
df2=pd.DataFrame(pps.matrix(df1))
df2[df2.case!='predict_itself'].iloc[:,[0,1,2,3,5,8]].sort_values('ppscore',ascending=False)
# ppscore is highest in petalwidth and species i.e, 0.927652 i.e, petal width is very good feature in predicting species
```
