### **univariate feature selection using chi square test**

```python
#out of 8 features of dataset which one is most imp in predicting the class?
# feature extraction with Unbivariate statistical Tests(chi-squared for classification)
from pandas import read_csv
import pandas as pd
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest # for univariate selction
from sklearn.feature_selection import chi2
```

```python
# load data, patients details taken, identify which variableis most imp in identifying whether patient has diabetes or not
filename='/content/pima-indians-diabetes.data.csv'
# in actual dataset no col names so give col names,Class-y variable
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(filename,names=names)
dataframe
```

```python
array=dataframe.values
array
```

```python
dataframe.shape
```

```python
x=array[:,0:8]
y=array[:,8]

# feature extraction
test=SelectKBest(score_func=chi2,k=5) # get top 5 features, K should be less than or equal to total no of features
fit=test.fit(x,y)
# summarize scores
set_printoptions(precision=4)
print(fit.scores_) # will get chi squares for each feature
# for regression: f_regression, mutual_info_regression
# for classification: chi2. f_classif,mutual_info_classif
# for feature 'test':chi square value is 2175.5653 so this is most imp feature,then plas,pedi...,
```

```python
names
```

```python
# score chi square values in a dataframe
x=pd.DataFrame(fit.scores_).T # .T will transpose list into dataframe
```

```python
x
```

```python
x.columns=names[:8]
x
```

```python
names
```

```python
# sort chi square values in ascending or descending values
sorted_x=x.sort_values(by=0,ascending=False,axis=1)
sorted_x

# axis=0:sort along the rows . i.e, you sort the dataframe based on the values in a specific column
# axis=1:sort along the columns . i.e, you sort the dataframe based on the values in a specific row
```

### **Recursive Feature Elimination**

```python
# feature extraction with RFE
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# load Data
filename='/content/pima-indians-diabetes.data.csv'

# in actual dataset no col names so give col names,Class-y variable
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(filename,names=names)
dataframe
```

```python
array=dataframe.values
array
```

```python
x=array[:,0:8]
y=array[:,8]
# feature extraction
model=LogisticRegression(max_iter=400) # to find out coefficients of logistic regression for that it need to iterate
# inside logistic regression optimization algorithm is there. using that, coefficients will be estimated
# for that it has to iterate multiple times.default value is 100
rfe=RFE(model) # see best features
fit=rfe.fit(x,y)
```

```python
# num Features:4 top features are available
fit.n_features_
```

```python
fit.support_
```

```python
names
```

```python
fit.ranking_
```

### Feature Importence using Decision Tree

```python
from pandas import read_csv
from sklearn.tree import DecisionTreeClassifier
# load Data
filename='/content/pima-indians-diabetes.data.csv'

# in actual dataset no col names so give col names,Class-y variable
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(filename,names=names)
dataframe
```

```python
array=dataframe.values
array
```

```python
x=array[:,0:8]
y=array[:,8]
# feature extraction
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
sorted_list=model.feature_importances_
print('feature importence',sorted_list)# before sorting
sorted_list.sort()
print('sorted feature importence',sorted_list) # after sorting
```
