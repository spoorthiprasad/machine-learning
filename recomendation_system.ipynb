```python
# change the structure of the dataset so that we can compute the similarity score
user_movies=movies.pivot(index="userId",
                         columns="movie",
                         values="rating")
```

```python
user_movies
# NaN-pereson didn't watch the movie so he didn't give rating. now we can recommend those (non watched ) movies to him
```

```python
user_movies.shape
```

```python
user_movies.fillna(0,inplace=True)
user_movies
```

```python
# calculating similarity b/w users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# user similarity- pairwise-taking two rows at a time it will compute distence
# that distence is scaled b/w 0 to 1
# Suppose distence b/w 2 rows is 0, where 0 indicates no distence and 2 indicates max distence,
# that is calculated as similarity=1-distence. so 1-0=1
# suppose distence is 0.9 so 1-0.9=0.1

user_sim=1-pairwise_distances(user_movies.values,metric="cosine")

# pairwise distences (..., metric='cosine) calculates the pairwise distences b/w rows in the matrix using
# 1-pairwise_distences(....) convert distences to similarities.
# cosine distences ranges from 0 to 2, where 0 indicates no distence and 2 indicates max distence
# so subtracting from 1 inverts this to give cosine similarity which ranges from -1 to 1


```

```python
user_sim  # this is similarity matrix- diagonal values are 1
```

```python
np.fill_diagonal(user_sim,0)
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index=movies.userId.unique()
user_sim_df.columns=movies.userId.unique()
```
